```python
import load
from tree import tmtree, treemodule
import tree.ctutils as ctu
import utils.sampling as smp
import tree.halomodule as hmo 
import numpy as np
import matplotlib.pyplot as plt
import tree
import draw
import pickle

is_gal = True
wdir = '/home/hoseung/Work/data/05427/'

# Load tree
alltrees = treemodule.CTree()

if is_gal:
    # Galaxy tree
    tree_path = 'GalaxyMaker/Trees/'
else:
    # halo tree
    tree_path = 'halo/Trees/'
    
try:
    alltrees = pickle.load(open(wdir + tree_path + "extended_tree.pickle", "rb" ))
    print("Loaded an extended tree")
except:
    alltrees = treemodule.CTree()
    alltrees.load(filename= wdir + tree_path + 'tree_0_0_0.dat')
    # Fix nout -----------------------------------------------------
    nout_max = alltrees.data['nout'].max()
    alltrees.data['nout'] += 187 - nout_max
    print("------ NOUT fixed")
    alltrees.data = ctu.augment_tree(alltrees.data, wdir, is_gal=is_gal)
    print("------ tree data extended")

nout = 186

info = load.info.Info(nout=nout, base=wdir, load=True)

hh = hmo.Halo(base=wdir, nout=nout, halofinder='HM', info=info, load=True, is_gal=is_gal)
```

```python
#for igal, gal in enumerate(final_gal):
#    for inout, nout in enumerate(nouts[1:]):
igal = 0
gal = 7
inout = 0
draw_part = True
snout = str(nout).zfill(3)

h = hmo.Halo(base=wdir, nout=nout, halofinder='HM', info=info, load=True, is_gal=True)
i_center = np.where(hh.data['np'] == max(hh.data['np']))

ind = smp.extract_halos_within(h.data, i_center, scale=3.0)
h_sub = tree.halomodule.Halo()
h_sub.derive_from(h, ind) 

region = smp.set_region(xc=h.data['x'][i_center],
                yc=h.data['y'][i_center],
                zc=h.data['z'][i_center], radius = 0.007)
                #radius = h.data['rvir'][i] * 10)


ptype=["star pos mass"]
#%%
```

```python
s = load.sim.Sim(nout, wdir, dmo=False, ranges=region["ranges"], setup=True)
s.add_part(ptype)
s.part.load()
part = getattr(s.part, s.part.pt[0])

x = part['x']
y = part['y']
z = part['z']
m = part['m'] * s.info.msun # part must be normalized already!
```

locations of halos on top of stellar density map. 

```python
from utils import util
util.reimport(draw)
util.reimport(draw.pp)


npix = 1200

imgs = draw.img_obj.MapSet(info=s.info, region=region)
imgp = draw.img_obj.MapImg(info=s.info, proj='z', npix=npix, ptype=ptype)
imgp.set_region(region)

imgp.set_data(draw.pp.den2d(x, y, z, m, npix, region=region, cic=True, norm_integer=True))
imgs.ptden2d = imgp
#    imgp.show_data()
#imgp.show_data()
for i in range(1,10):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)

    imgp.plot_2d_den(ax=ax1, show=False,zposition=False)
    ind = np.where((h.data['m'] > 10**(9 + i/3.)) * (h.data['m'] < 10**(9 + (i+1)/3.)))
    h_sub = tree.halomodule.Halo()
    h_sub.derive_from(hh, ind) 

    #ax1.set_xlim(right=npix).
    #ax1.set_ylim(top=npix)
    draw.pp.pp_halo(h_sub, npix, region=region, axes=ax1, rscale=4, name=True, radius="r",\
                    fontsize=3, linewidth=0.3) # Use galaxy radius rather than virial radius
    plt.savefig(wdir+"gal_and_halo_" + str(i) + ".png", dpi=300)
    plt.clf() #clear figure
```
