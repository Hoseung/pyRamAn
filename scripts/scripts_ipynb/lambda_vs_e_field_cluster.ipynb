```python
def ind_cluster_field(cat, halo, info, rscale = 2.0):
    """
    Assume that the most massive galaxy is the BCG
    """
    cluster = halo.data[np.argmax(halo.data['np'])]
    clu_x = cluster['x'] * info.pboxsize
    clu_y = cluster['y'] * info.pboxsize
    clu_z = cluster['z'] * info.pboxsize
    rcluster = cluster['rvir'] * info.pboxsize
    dist = np.sqrt(np.square(clu_x - cat['xc']) + 
                   np.square(clu_y - cat['yc']) + 
                   np.square(clu_z - cat['zc']))
        
    i_clu = dist < (rcluster * rscale)
    i_field = dist > (rcluster * rscale)
    return i_clu, i_field


def pp_density_map(axmain, x, y,
                    xmin=-0.05, xmax=0.9,
                    ymin=-0.05, ymax=0.9,
                    levels=None, color=True, cmap="winter"):
    # Draw main density map 
    xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([xx.ravel(), yy.ravel()])

    values = np.vstack([x, y])
    kernel = st.gaussian_kde(values)
    f = np.reshape(kernel(positions).T, xx.shape)
    # , [0.2, 2, 3, 5, 10, 16]  custom contour levels.
    f /= max(f.ravel())
    
    cfset = axmain.contour(xx, yy, f, levels=levels, cmap=cmap, lw=5)
    return cfset


def plot_sami(ax, data, contour=True, scatter=False, levels=None):
    x = data['ellp']
    y = data['r1']

    xmin, xmax = -0.05, 0.9
    ymin, ymax = -0.05, 0.9

    if contour:
        plot_densiy_map(ax, x, y, xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax,
                        levels=levels, color=False, cmap="spring")
    if scatter:
        ax.scatter()
        
        
        
class params():
    def __init__(self):
        pass
    
    
class lambda_data():
    
    def __init__(self, x, y, ids, name="name", sfc=None, sec=None, salph=1,
                 scmap="jet", ccmap=None, hec=None):
        self.edata = x
        self.ldata = y
        self.ids = ids
        self.data_name = name
        
        self.do_scatter = False           
        
        self.scatter = params()
        self.contour = params()
        self.hist = params()
        
        # Default values
        # If no color is specified, it is determined by matplotlib automatically.
        self.scatter.color = False
        self.scatter.fc = sfc
        self.scatter.ec = sec
        self.scatter.alpha = salph
        self.scatter.cmap = scmap
        
        
        
        self.contour.cmap = ccmap
        
        
        self.hist.xec = hec
        self.hist.yec = hec
        self.hist.binsize = 0.05
        self.hist.nbin_x = np.floor((max(x) - min(x)) / self.hist.binsize)
        self.hist.nbin_y = np.floor((max(y) - min(y)) / self.hist.binsize)
        self.hist.range_x = [0, 0.8]
        self.hist.range_y = [0, 0.8]
        
        
    
# Draw main density map 
def plot_density_map(datasets, fig, levels = [0.1, 0.6, 0.95], kwargs={}):
    """
        plot density map of multiple data sets.
        
        Only contours are plotted by default 
        because scatter plot of multiple data sets is too busy. 
        
        Scattered points may be color-coded, but color-coding of multiple data sets 
        again makes no sense. Dont' use color coding unless only one data set is plotted.
        (multiple) colorbars for the color-coding will make the plot even crazier.
        
        
    """
    axmain = fig.axes[0]
    axHistX = fig.axes[1]
    axHistY = fig.axes[2]
    
    # legends
    lines = []
    labels = []
    for data in datasets:
        
        xdata, ydata, ids = data.edata, data.ldata, data.ids
        scatter = data.scatter
        contour = data.contour
        hist = data.hist
        
        
        if data.do_scatter:
            if data.color:
                c = np.log10(stellarmass) #rank
                sct = axmain.scatter(xdata,ydata, s=20, c=c, cmap=scatter.cmap,
                         alpha=scatter.alpah,
                         facecolor=scatter.fc, 
                         edgecolor=scatter.ec)
                ains=plt.axes([0.65,0.4,0.05,0.3])
                cbar = plt.colorbar(sct, cax=ains, orientation='vertical')
                cbar.ax.yaxis.set_ticks_position('left')
                cbar.ax.set_ylabel(r"$log10(M_{*})$")
                cbar.set_ticks(np.arange(0, 0.95, 0.15))
                cbar.set_ticks(levels)
            else:
                sct = axmain.scatter(xdata,ydata, s=20, 
                                         facecolor=scatter.fc, edgecolor=scatter.ec)
                cbar = None


        #levels = None
        

        acontour = pp_density_map(axmain, xdata, ydata,
                                levels=levels,
                                cmap = contour.cmap)
        axmain.clabel(acontour, inline=1, fontsize=10)

        lines.append(acontour.collections[0])
        labels.append(data.data_name)

        if annotate:
            print("annotating")
            for a,b,id in zip(xdata,ydata,ids):
                axmain.annotate(str(id),xy=(a,b))


        # Draw x-axis histogram
        hh = axHistX.hist(xdata, hist.nbin_x, ec=hist.xec, fc='none', histtype='step')
        #hh = axHistX.hist(x[i_field], num_x_bins, ec='b', fc='none', histtype='step')

        # Draw y-axis histogram
        hh = axHistY.hist(ydata, hist.nbin_y, ec=hist.yec, fc='none', histtype='step',
                     orientation='horizontal')
        #hh = axHistY.hist(y[i_field], num_y_bins, ec='b', fc='none', histtype='step',
        #             orientation='horizontal')

        

    # add some margin to histograms.
    # reduce number of ticks
    axHistX.set_ylim([0,1.1 * max(hh[0])])
    axHistX.locator_params(nbins=4)        
        
    axHistY.set_xlim([0,1.1 * max(hh[0])])
    axHistY.locator_params(nbins=4)
    
    axmain.legend(lines, labels, shadow=True, fancybox=True)
    
    
    
# Prepare figure ####################################################

def base_figure(xmin = -0.05, xmax = 0.9, ymin = -0.05, ymax = 0.9,
                binsize = 0.05):
    fig = plt.figure(221, figsize=(6.3, 6.3))

    # Calculate number of bins based on binsize for both x and y
    min_x_data, max_x_data = 0, xmax
    num_x_bins = np.floor((max_x_data - min_x_data) / binsize)

    min_y_data, max_y_data = 0, ymax
    num_y_bins = np.floor((max_y_data - min_y_data) / binsize)

    # Axes definitions
    nullfmt = plt.NullFormatter()
    hist_height = 0.2
    left, width = 0.16, 0.97 - hist_height - 0.15# 0.6
    bottom, height = 0.1, 0.97 - hist_height - 0.15# 0.6
    left_h = left + width + 0.01
    bottom_h = bottom + height + 0.01

    rect_main = [left, bottom, width, height]
    rect_histx = [left, bottom_h, width, hist_height]
    rect_histy = [left_h, bottom, hist_height, height]

    axmain=fig.add_subplot(223, position=rect_main)
    axmain.set_xlim(xmin, xmax)
    axmain.set_ylim(ymin, ymax)
    # suppress last tick
    axmain.set_xticks(np.arange(0, xmax, 0.1))
    #axmain.set_xticklabels(['0.5','1.0'])
    axmain.set_yticks(np.arange(0, ymax, 0.1))
    axmain.set_xlabel(r"$\epsilon_{R_{e}}$", fontsize=16)
    axmain.set_ylabel(r"$\lambda_{R_{e}}$", fontsize=16)

    # S/R demarcation line
    axmain.plot(sr_line_xx, sr_line_yy, '--', lw=1, color='black')

    axHistX = fig.add_subplot(221, position=rect_histx)
    axHistX.set_ylabel("# of galaxies")
    axHistY = fig.add_subplot(221, position=rect_histy)
    axHistY.set_xlabel("# of galaxies")

    # Remove labels from histogram edges touching scatter plot
    axHistX.xaxis.set_major_formatter(nullfmt)
    axHistY.yaxis.set_major_formatter(nullfmt)
    
    return fig
```
